### Task 5: Distribution of Languages in Film

```python
language_data = train.loc[train['original_language'].isin(train['original_language'].value_counts().head(10).index)]
```

```python
plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
sns.boxplot(x='original_language', y='revenue',data=language_data)
plt.title('Mean revenue per language')
plt.subplot(1,2,2)
sns.boxplot(x='original_language', y='log_revenue',data=language_data)
plt.title('Mean log revenue per language')
```

```python
# Per inspeciton from the boxplot we could identify that the English movies were able to generate maximum revenues for most of the time. But it might be biased from log transformation as other languages can also generate high revenue. 
```

### Task 6: Frequent Words in Film Titles and Descriptions

```python
#display the top words shown from mivie titles
sw = stopwords.words("english")
plt.figure(figsize=(12,12))
text = ' '.join(word for word in train['original_title'].values if word not in sw)
wordcloud = WordCloud(max_font_size=None,
                      background_color='white',
                      width=1200,height=1000).generate(text)
plt.imshow(wordcloud)
plt.title('Top words across movie titles')
plt.axis('off')
plt.show()                     
```

```python
#Now move to examine top words query on the movie descriptions
train['overview'].fillna('',inplace=True) # filter off the null value from descriptions

plt.figure(figsize=(12,12))
text = ' '.join(word for word in train['overview'].values if word not in sw)
wordcloud = WordCloud(max_font_size=None,
                      background_color='white',
                      width=1200,height=1000).generate(text)
plt.imshow(wordcloud)
plt.title('Top words across movie descriptions')
plt.axis('off')
plt.show()    
```

### Task 7: Do Film Descriptions Impact Revenue?

```python
import eli5
from sklearn.linear_model import LinearRegression
```

```python
vectorizer = TfidfVectorizer(
    sublinear_tf=True,
    analyzer='word',
    token_pattern=r'\w{1,}',
    ngram_range=(1, 2),
    min_df=5
)
overview_text = vectorizer.fit_transform(word for word in train['overview'].fillna('') if word not in ['to'])
lin_model = LinearRegression()
lin_model.fit(overview_text,train['log_revenue'])
eli5.show_weights(lin_model,vec=vectorizer,top=20,feature_filter= lambda x:x!='<BIAS>')
```
